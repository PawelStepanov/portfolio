```python
def count_missing(data):
    '''
    The function counts the number of missing values
    in a dataframe, as well as their share relative 
    to all values. Creates a dataframe with two columns. 
    The first is Missing, the second is Share
    args:
        data (DataFrame): the dataframe in which missing 
                          values need to be calculated
    returns:
        data (DataFrame): final dataframe
    '''

    count_na = pd.Series(data.isna().sum(), name='missing')
    proport_na = pd.Series(
        round((data.isna().sum()/len(data)) * 100, 2), name='proportion')

    return pd.concat([count_na, proport_na], axis=1).sort_values(by='proportion', ascending=False)
```

```python
def test_normal(data):
    '''
    The function tests the data for normal distribution
    using the D'Agostino test. Takes a column and compares
    its p-value with the statistical significance level, 
    outputs a response about normality
    args:
        data (DataFrame): dataframe column
        alpha (float): significance level
    return:
        None
    '''
    
    stat, p = normaltest(data)
    
    if p > ALPHA:
        print('We do not reject H0. The data is from a normal distribution')
    else:
        print('We reject H0 and accept H1. \
The data is not from a normal distribution.')
```

- **H0** - The data is obtained from a normal distribution

<div style="background-color:#f9f9f9; border-left:5px solid #40407A; padding:10px; font-family:Verdana, sans-serif; color:#333;">
    This section shows the results of the best attempt, and also shows the processes by which I obtained the hyperparameters of the model for my best attempt.

To determine the hyperparameters, I used optuna and made predictions on the iced dataset (data that was encoded in the normal way + data encoded using RAPIDS cuML Target Encoder)
    
Since the EDA input showed that our data is very noisy and there is almost no signal with the target, all because the data was synthesized based on a synthesized data set. Therefore, to improve the result, it is necessary to catch a weak signal in the data and for this, it was necessary to create many new features.
</div>

```python
def encoder(df):
    '''
    The function produces encoding of features (Ordinal, Ohe, Standard)
    args:
        df (DataFrame): original dataframe
    returns:
        data (DataFrame): final dataframe
    '''

    data = df.copy()

    # Ordinal encoding
    data['Size'] = data['Size'].replace({'Unknown': 1, 'Small': 2, 'Medium': 3, 'Large': 4})
    data['Weight_Class'] = data['Weight_Class'].replace({'Light': 1, 'Middle': 2, 'Light_heavy': 3,
                                                        'Middle_heavy': 4, 'Heavy': 5})
    
    # OneHot encoding
    data = pd.get_dummies(data, columns=['Material', 'Style', 'Color', 'Brand', 'Waterproof', 'Laptop Compartment'], 
                          prefix=['Material', 'Style', 'Color', 'Brand', 'Waterproof', 'Laptop Compartment'], 
                          drop_first=True, dtype='int')
    
    # Standard scaler   
    data['Weight Capacity (kg)'] = (data['Weight Capacity (kg)'] -
                                    data['Weight Capacity (kg)'].mean()) / data['Weight Capacity (kg)'].std()

    return data
```

```python
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        "max_depth": trial.suggest_int("max_depth", 2, 10),
        "num_leaves": trial.suggest_int("num_leaves", 50, 250),
        "learning_rate": trial.suggest_float("learning_rate", 1e-3, 0.1),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 100),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 1),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 1)
    }
    
    model = LGBMRegressor(**params, verbose=-1, random_state=RANDOM_STATE)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    
    return root_mean_squared_error(y_valid, y_pred)
```

<div style="background-color:#f9f9f9; border-left:5px solid #40407A; padding:10px; font-family:Verdana, sans-serif; color:#333;">
We will check all possible combinations of features and determine the best ones, so that we can use them in the future.
    
</div>

```python
%%time

# Categorical features
cat_features = ['Brand', 'Material', 'Size', 'Compartments', 'Laptop Compartment', 'Waterproof', 'Style', 'Color']

# Maximum number of features in a combination
max_features = 6

# List to save results
results = []

# Searching for all combinations from 2 to max_features
for r in range(2, max_features+1):
    for combo in combinations(cat_features, r):
        # Create a new column title
        new_col_name = '_'.join(combo)

        train_new_col = train_total.copy()
        
        # Combining features
        train_new_col[new_col_name] = train_new_col[list(combo)].astype(str).agg('_'.join, axis=1)
        
        # We encode categorical features
        X = train_new_col[new_col_name].factorize()[0].reshape(-1, 1)
        y = train_new_col['Price']  # Target variable
        
        # We divide into train and test
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
        
        # Training the model
        model = Ridge()
        model.fit(X_train, y_train)
        
        # We evaluate the result
        y_pred = model.predict(X_test)
        score = mean_squared_error(y_test, y_pred)
        
        # We save the combination and its evaluation
        results.append((combo, score))
        print(f"Combination: {combo} with MSE: {score}")

# Sort by score (MSE) and select top 20
top_20_combos = sorted(results, key=lambda x: x[1])[:20]

print("\nTop-20 Combinations:")
for combo, score in top_20_combos:
    print(f"Combination: {combo}, MSE: {score}")
```

<div style="background-color:#f9f9f9; border-left:5px solid #40407A; padding:10px; font-family:Verdana, sans-serif; color:#333;">   
    Next we will design features using groupby(COL1)[COL2].agg(STAT). This is where we group by the COL1 columns and aggregate (i.e. calculate) the STAT statistics on another column COL2. This is the basic method used to calculate target encoding and count encoding. By calculating the raw statistics groupby(COL1)[COL2].agg(STAT) and feeding it into our model.
</div>

```python
CATS = list(train_total.columns[1:-2])
print(f"There are {len(CATS)} categorical columns:")
print( CATS )
print(f"There are 1 numerical column:")
print( ["Weight Capacity (kg)"] )
```

```python
for list_col, _ in top_20_combos[:10]:
    new_col_name = '_'.join(list_col)
    train_total[new_col_name] = train_total[list(list_col)].astype('str').agg('_'.join, axis=1)
    test_total[new_col_name] = test_total[list(list_col)].astype('str').agg('_'.join, axis=1)
    CATS.append(new_col_name)
```

```python
COMBO = []
for i,c in enumerate(CATS):
    combine = pd.concat([train_total[c],test_total[c]],axis=0)
    combine,_ = pd.factorize(combine)
    train_total[c] = combine[:len(train_total)]
    test_total[c] = combine[len(train_total):]
    n = f"{c}_wc"
    train_total[n] = train_total[c]*100 + train_total["Weight Capacity (kg)"]
    test_total[n] = test_total[c]*100 + test_total["Weight Capacity (kg)"]
    COMBO.append(n)
print()
print(f"We engineer {len(COMBO)} new columns!")
print( COMBO )
```

```python
FEATURES = CATS + ["Weight Capacity (kg)"] + COMBO
print(f"We now have {len(FEATURES)} columns:")
print( FEATURES )
```

```python
# STATISTICS TO AGGEGATE FOR OUR FEATURE GROUPS
STATS = ["mean","std","count","nunique","median","min","max","skew", "var", "sem"]
STATS2 = ["mean","std", "var", "median"]
```

```python
col = train_total.columns.to_list()

for c in col[:9]:
    train_total[c] = train_total[c].astype('int32')
    test_total[c] = test_total[c].astype('int32')

for c in col[11:21]:
    train_total[c] = train_total[c].astype('int32')
    test_total[c] = test_total[c].astype('int32')

for c in col[21:]:
    train_total[c] = train_total[c].astype('float32')
    test_total[c] = test_total[c].astype('float32')
```

```python
lgbm_params = {'n_estimators': 451, 'max_depth': 8, 'num_leaves': 233, 'learning_rate': 0.03000044687948457, 
               'min_child_samples': 62, 'subsample': 0.9694010630437035, 'colsample_bytree': 0.5645494636160868, 
               'reg_alpha': 0.564820643502276, 'reg_lambda': 0.4346072158112042}
```

```python
%%time

FOLDS = 7
kf = KFold(n_splits=FOLDS, shuffle=True, random_state=42)

oof_train = np.zeros((train_total.shape[0], ))
pred = np.zeros((test_total.shape[0], ))
oof_test_skf = np.empty((FOLDS, test_total.shape[0]))

# OUTER K FOLD
for i, (train_index, valid_index) in enumerate(kf.split(train_total)):
    print(f"### OUTER Fold {i+1} ###")

    X_train = train_total.loc[train_index,FEATURES+['Price']].reset_index(drop=True).copy()
    y_train = train_total.loc[train_index,'Price']

    X_valid = train_total.loc[valid_index,FEATURES].reset_index(drop=True).copy()
    y_valid = train_total.loc[valid_index,'Price']

    X_test = test_total[FEATURES].reset_index(drop=True).copy()

    # INNER K FOLD (TO PREVENT LEAKAGE WHEN USING PRICE)
    kf2 = KFold(n_splits=FOLDS, shuffle=True, random_state=42)   
    for j, (train_index2, valid_index2) in enumerate(kf2.split(X_train)):
        print(f" ## INNER Fold {j+1} (outer fold {i+1}) ##")

        X_train2 = X_train.loc[train_index2,FEATURES+['Price']].copy()
        X_valid2 = X_train.loc[valid_index2,FEATURES].copy()

        ### FEATURE SET 1 (uses price) ###
        col = "Weight Capacity (kg)"
        tmp = X_train2.groupby(col).Price.agg(STATS)
        tmp.columns = [f"TE1_wc_{s}" for s in STATS]
        X_valid2 = X_valid2.merge(tmp, on=col, how="left")
        for c in tmp.columns:
            X_train.loc[valid_index2,c] = X_valid2[c].values

        ### FEATURE SET 2 (uses price) ###
        for col in COMBO:
            tmp = X_train2.groupby(col).Price.agg(STATS2)
            tmp.columns = [f"TE2_{col}_{s}" for s in STATS2]
            X_valid2 = X_valid2.merge(tmp, on=col, how="left")
            for c in tmp.columns:
                X_train.loc[valid_index2,c] = X_valid2[c].values

    ### FEATURE SET 1 (uses price) ###
    col = "Weight Capacity (kg)"
    tmp = X_train.groupby(col).Price.agg(STATS)
    tmp.columns = [f"TE1_wc_{s}" for s in STATS]
    X_valid = X_valid.merge(tmp, on=col, how="left")
    X_test = X_test.merge(tmp, on=col, how="left")
    
    ### FEATURE SET 2 (uses price) ###
    for col in COMBO:
        tmp = X_train.groupby(col).Price.agg(STATS2)
        tmp.columns = [f"TE2_{col}_{s}" for s in STATS2]
        X_valid = X_valid.merge(tmp, on=col, how="left")
        X_test = X_test.merge(tmp, on=col, how="left")

    ### FEATURE SET 3 (does not use price) ###
    for col in CATS:
        col2 = "Weight Capacity (kg)"
        tmp = X_train.groupby(col)[col2].agg(STATS2)
        tmp.columns = [f"FE3_{col}_wc_{s}" for s in STATS2]
        X_train = X_train.merge(tmp, on=col, how="left")
        X_valid = X_valid.merge(tmp, on=col, how="left")
        X_test = X_test.merge(tmp, on=col, how="left")

    # DROP PRICE THAT WAS USED FOR TARGET ENCODING
    X_train = X_train.drop(['Price'],axis=1)

    # BUILD MODEL
    model = LGBMRegressor(**lgbm_params, n_jobs=-1, random_state=RANDOM_STATE, verbose=-1)
    
    # TRAIN MODEL
    COLS = X_train.columns
    model.fit(
        X_train[COLS], y_train,
        eval_set=[(X_valid[COLS], y_valid)],  
        callbacks=[log_evaluation(0), early_stopping(100, verbose=False)])

    # PREDICT OOF AND TEST
    oof_train[valid_index] = model.predict(X_valid[COLS])
    oof_test_skf[i, :] = model.predict(X_test[COLS])

pred[:] = oof_test_skf.mean(axis=0)
```

```python
# COMPUTE OVERALL CV SCORE
true = train_total.Price.values
s = np.sqrt(np.mean((oof_train-true)**2.0))
print(f"=> Overall CV Score = {s}")
```
