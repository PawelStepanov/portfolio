```python
def count_missing(data):
    '''
    The function counts the number of missing values
    in a dataframe, as well as their share relative 
    to all values. Creates a dataframe with two columns. 
    The first is Missing, the second is Share
    args:
        data (DataFrame): the dataframe in which missing 
                          values need to be calculated
    returns:
        data (DataFrame): final dataframe
    '''
    count_na = pd.Series(data.isna().sum(), name='missing')
    proportion_na = pd.Series(
        (data.isna().sum() / len(data)) * 100, name='proportion_missing')
    return pd.concat([count_na, proportion_na], axis=1).sort_values(
        by='proportion_missing', ascending=False)
```

```python
def test_normal(data, alpha=0.05):
    '''
    The function tests the data for normal distribution
    using the D'Agostino test. Takes a column and compares
    its p-value with the statistical significance level, 
    outputs a response about normality
    args:
        data (DataFrame): dataframe column
        alpha (float): significance level
    return:
        None
    '''

    stat, p = normaltest(data)
    
    if p > alpha:
        print('We do not reject H0. The data is from a normal distribution')
    else:
        print('We reject H0 and accept H1. \
The data is not from a normal distribution.')
```

```python
def plot_cat(df, col):
    '''
    The function plots two graphs: pie and barplot
    args:
        df (DataFrame): origingal dataframe
        col (str): name of column
    returns:
        None
    '''
    
    data = df[col].value_counts()
    
    plt.figure(figsize=(8, 4))
    
    plt.subplot(1, 2, 1)
    plt.pie(data.values, labels=data.index, autopct='%.1f%%')
    plt.title(f'Pie chart of {col}')
    
    plt.subplot(1, 2, 2)
    data.plot(kind='bar', color=mcolors.TABLEAU_COLORS)
    plt.ylabel('Count')
    plt.title(f'Barplot of {col}')
```

    We will fill in the missing values as separate categories. Later, during the discussion on kaggle, it turned out that this was the right decision.

```python
def filling_missing(df):
    '''
    The function takes a dataframe and fills in the missing values in each column.
    For numeric data, this is -1, for categorical data - Unknown.
    args: 
        df (DataFrame): original dataframe
    returns:
        df (DataFrame): final datagrame
    '''
    
    df['Insurance Duration'] = df['Insurance Duration'].fillna(-1)
    
    df['Vehicle Age'] = df['Vehicle Age'].fillna(-1)

    df['Marital Status'] = df['Marital Status'].fillna('Unknown')
    
    df['Age'] = df['Age'].fillna(-1)

    df['Annual Income'] = df['Annual Income'].fillna(-1)

    df['Health Score'] = df['Health Score'].fillna(-1)

    df['Customer Feedback'] = df['Customer Feedback'].fillna('Unknown')

    df['Number of Dependents'] = df['Number of Dependents'].fillna(-1)

    df['Credit Score'] = df['Credit Score'].fillna(-1)

    df['Occupation'] = df['Occupation'].fillna('Unknown')

    df['Previous Claims'] = df['Previous Claims'].fillna(-1)
    
    return df
```

```python
def feature_enginer(df):
    '''
    The function creates new features based on Policy Start Year.
    First, the year, month and day are taken from the start date.
    Then, harmonic encoding is used to account for their periodicity.
    args:
        df (DataFrame): original dataframe
    returns:
        df (DataFrame): final dataframe
    '''
    
    df['Policy Start Year'] = pd.to_datetime(df['Policy Start Date']).dt.year
    df['Policy Start month'] = pd.to_datetime(df['Policy Start Date']).dt.month
    df['Policy Start day'] = pd.to_datetime(df['Policy Start Date']).dt.day
    #df['Policy Start dayofweek'] = pd.to_datetime(df['Policy Start Date']).dt.dayofweek
    #df['Month_name'] = df['Policy Start Date'].dt.month_name()
    #df['Day_of_week'] = df['Policy Start Date'].dt.day_name()
    #df['Week'] = df['Policy Start Date'].dt.isocalendar().week
    
    # Harmonic encoding
    df['Year_sin'] = np.sin(2 * np.pi * df['Policy Start Year'])
    df['Year_cos'] = np.cos(2 * np.pi * df['Policy Start Year'])
    df['Month_sin'] = np.sin(2 * np.pi * df['Policy Start month'] / 12)
    df['Month_cos'] = np.cos(2 * np.pi * df['Policy Start month'] / 12)
    df['Day_sin'] = np.sin(2 * np.pi * df['Policy Start day'] / 31)
    df['Day_cos'] = np.cos(2 * np.pi * df['Policy Start day'] / 31)

    # Grouping by time
    #df['Group'] = (df['Policy Start Year'] - 2020) * 48 + df['Policy Start month'] * 4 + df['Policy Start day'] // 7    
    
    #df['vehicle_age_insurance'] = df['Vehicle Age'] / df['Insurance Duration']
    #df['annual_income_health'] = df['Annual Income'] / df['Health Score']
    #df['annual_income_age'] = df['Annual Income'] / df['Age']
    #df['health_score_age'] = df['Health Score'] / df['Age']
    
    df = df.drop(columns=['Policy Start Date', 'Policy Start Year', 'Policy Start month', 'Policy Start day'], axis=1)
    
    return df
```

```python
def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimatros", 500, 2000),
        "max_depth": trial.suggest_int("max_depth", 5, 20),
        "num_leaves": trial.suggest_int("num_leaves", 50, 350),
        "learning_rate": trial.suggest_float("learning_rate", 1e-3, 0.1),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 100),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 10.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 10.0)
    }
    
    model = LGBMRegressor(**params, verbose=-1, random_state=RANDOM_STATE)
    model.fit(X_train_prep, y_train_log)
    y_pred = model.predict(X_valid_prep)
    
    return rmsle(np.expm1(y_valid_log), np.expm1(y_pred))
```

```python
mlp_col = []
for i in num_col:
    mlp_col.append(f'num__{i}')
```

```python
# Initialize the models
ridge = Ridge(alpha=1.0)
lgbm = LGBMRegressor(n_jobs=-1, verbose=-1, random_state=RANDOM_STATE)
catboost = CatBoostRegressor(verbose=0, random_state=RANDOM_STATE)
mlp = MLPRegressor(random_state=RANDOM_STATE)
```

```python
def get_oof_predictions(model, X_train, y_train, X_test, n_splits=5, cb=False):
    '''
    Generating out-of-fold predictions for training a meta-model.
    '''
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=RANDOM_STATE)
    
    global cat_col
    
    oof_train = np.zeros((X_train.shape[0],))
    oof_test = np.zeros((X_test.shape[0],))
    oof_test_skf = np.empty((n_splits, X_test.shape[0]))
    
    for fold, (train_idx, valid_idx) in enumerate(kf.split(X_train)):
        if isinstance(X_train, np.ndarray):
            X_tr, X_val = X_train[train_idx], X_train[valid_idx]
        else:
            X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[valid_idx]
            
        y_tr, y_val = y_train[train_idx], y_train[valid_idx]
        
        if cb:
            model.fit(X_tr, y_tr, cat_features=cat_col)
            oof_train[valid_idx] = model.predict(X_val)
            oof_test_skf[fold, :] = model.predict(X_test)
        else:
            model.fit(X_tr, y_tr)
            oof_train[valid_idx] = model.predict(X_val)
            oof_test_skf[fold, :] = model.predict(X_test)            
    
    # Average fold predictions
    oof_test[:] = oof_test_skf.mean(axis=0)
    
    return oof_train, oof_test
```

```python
# Predictions on Ridge
ridge_oof_train, ridge_oof_test = get_oof_predictions(
    ridge, X_train_prep, y_train_log.values, df_test_prep)
```

```python
# Predictions on LGBMRegressor
lgbm_oof_train, lgbm_oof_test = get_oof_predictions(
    lgbm, X_train_prep, y_train_log.values, df_test_prep)
```

```python
# Selecting categorical features
cat_col = X.select_dtypes(include='object').columns.to_list()
```
